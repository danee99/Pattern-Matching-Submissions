```python
# Pattern Matching and Machine Learning for Audio Signal Processing
# Sheet 3, Exercise 3
# Martin Br√ºckmann, Daniel Degenstein, Raphael Wude, Claude Jordan

import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
# Calculates the N-th roots of unity
# with the formula exp(2 * k * pi * i / N)
# for k=0,1,...,N-1
def calculate_Nth_roots_of_unity(N):
    root_list = []
    for k in range(0,N):
        z = np.exp(2* np.pi * 1j*k/N)
        root_list.append(z)
    return root_list

# Plots the roots of unity calculated with above function
def plot_Nth_root_of_unity(N):
    root_list = calculate_Nth_roots_of_unity(N)
    for z in root_list:
        plt.plot([0,np.real(z)], [0, np.imag(z)])
    plt.axis("square")
    x = np.linspace(0, 2*np.pi, 100)
    plt.plot(np.cos(x), np.sin(x), color="gray")
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("x")
    plt.ylabel("y")
    #plt.savefig("6th_root_of_unity.png")
    plt.show()
    
plot_Nth_root_of_unity(6)
```

```python
# Calculates primitive roots of unity
# Does this by checking, for any given calculated root of unity,
# whether this is also a root of unity for m=1,...,N-1
# If it is, then the root is discarded because it is not a primitve root
def calculate_primitive_Nth_root_of_unity(N):
    root_list = []
    for k in range(0,N):
        append_z = True
        z = np.exp(2* np.pi * 1j*k/N)
        for m in range(0,N):
            previous_root_list = calculate_Nth_roots_of_unity(m)
            if z in previous_root_list:
                append_z = False
        if append_z:
            root_list.append(z)
    return root_list

# Plots the primitive roots of unity
# Same procedure as for regular roots of unity
def plot_primitive_Nth_root_of_unity(N):
    primitive_root_list = calculate_primitive_Nth_root_of_unity(N)
    for z in primitive_root_list:
        plt.plot([0,np.real(z)], [0, np.imag(z)])
    plt.axis("square")
    x = np.linspace(0, 2*np.pi, 100)
    plt.plot(np.cos(x), np.sin(x), color="gray")
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    plt.xlabel("x")
    plt.ylabel("y")
    #plt.savefig("primitive_6th_root_of_unity.png")
    plt.show()
    
plot_primitive_Nth_root_of_unity(6)
```
